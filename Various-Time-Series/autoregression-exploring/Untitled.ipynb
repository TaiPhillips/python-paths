### Simple Autoregression models

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
y_start = [21,22]
y = y_start
for i in range(100):
    y_next = 8 + 1.3 * y[i+1] - .7 * y[i] + np.random.normal()
    y.append(y_next)
plt.plot(np.arange(len(y)),y)
plt.show()
```

```python
y_start = [11]
y = y_start
for i in range(100):
    y_next = 18 - 0.8 * y[i] + np.random.normal()
    y.append(y_next)
plt.plot(np.arange(len(y)),y)
plt.show()
```

```python
y_start = np.array([21,22])
y1 = y_start.copy()
y2 = y_start.copy()
for i in range(100):
    y_next1 = 8 + 1.3 * y1[i+1] - .7 * y1[i] + np.random.normal(scale=0.2)
    y_next2 = 8 + 1.3 * y2[i+1] - .7 * y2[i]
    y1 = np.append(y1, [y_next1])
    y2 = np.append(y2, [y_next2])

fig, axes = plt.subplots(2,1, figsize=(13,7))
axes[0].plot(np.arange(len(y1)),y1)
axes[1].plot(np.arange(len(y2)),y2)
plt.show()
```
